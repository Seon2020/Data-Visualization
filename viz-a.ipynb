```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
cars = pd.read_csv('Car_emission_sample.csv')
cars.head()
```

```python
sns.relplot(x='FUEL_CONSUMPTION*', y='CO2_EMISSIONS', hue='VEHICLE CLASS', data=cars)
```

- We can clearly see above that there is a clear relationship between a vehicle's fuel consumption and CO2 emissions.

```python
sns.lineplot(x='FUEL_CONSUMPTION*', y='CO2_EMISSIONS', data=cars, ci=None)

```

- As you can see in the above line plot, we can infer that the more fuel a vehicle consumes to fill the tank, the most CO2 emissions that vehicle produces. We can see that the relationship is not as strong with vehicles that consume less than ~9 and vehicles that consume more than ~19. The relationship is strongest for vehicles that consume anywhere between 9 and 19 to get a full tank.

```python
group_by_make = cars.groupby('MAKE', as_index=False).sum()
sns.lmplot(x='ENGINE_SIZE', y='CO2_EMISSIONS', data=group_by_make)
```

- We can see above that after grouping cars by their 'MAKE', the points are still aligned linearly. This tells us that the make does not significantly affect the relationship between engine size and CO2 emissions. 

```python
sns.boxplot(x='ENGINE_SIZE', y='CO2_EMISSIONS', data=cars)
```

- We can see in the boxplot above that outliers are more likely when a vehicle's engine size is between 1.7 and 2.5. The rest of the outliers in this data set are few and far between. This tells us that the relationship between engine size and CO2 emissions is weakest when the engine size is smaller. 

### Conclusions:
- We can conclude that there is a linear relationship between fuel consumption and CO2 emissions. 
- The relationship between fuel consumption and CO2 emissions is strongest for vehicles that consume anywhere between 9 and 19 to get a full tank.
- We can conclude that there is a strong relationship between engine size and CO2 emissions.
- The relationship between engine size and CO2 emissions tends to be weaker when the engine size is smaller
